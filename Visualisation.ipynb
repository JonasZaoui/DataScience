```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

# Pairplot: la vue d'ensemble

```python
sns.pairplot(iris, hue='species')
```

# Visualisation de catégories 

```python
titanic = sns.load_dataset('titanic')
titanic.drop(['alone', 'alive', 'who', 'adult_male', 'embark_town', 'class'], axis=1, inplace=True)
titanic.dropna(axis=0, inplace=True)
titanic.head()
```

```python
sns.catplot(x='survived', y='age', data=titanic, hue='sex')
```

```python
sns.boxplot(x='pclass', y='age', data=titanic, hue='sex')
```

Au milieu de chaque boxplot, on retrouve la médiane. Sur la premiere boxplot, la médiane des ages des hommes en premiere classe est 40 ans. 
1er quartile : 30 ans  (25% ont moins de 30 ans)
2e quartile : 50 ans.

Pour la deuxieme boxplot bleues, il y a deux outliers.

```python
plt.figure(figsize=(32, 8))
sns.boxplot(x='age', y='fare', data=titanic, hue='sex')
```

# Visualisation de Distributions


```python
sns.distplot(titanic['fare'])
```

```python
sns.jointplot('age', 'fare', data=titanic, kind='hex') #kind = style 
```

```python
sns.heatmap(titanic.corr()) #affichage des corrélations entre les variables 
```
